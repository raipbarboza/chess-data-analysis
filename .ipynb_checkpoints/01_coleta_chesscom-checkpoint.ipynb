```python
import requests
import pandas as pd
import json
from datetime import datetime

USERNAME = "raizeraaa"
headers = {"User-Agent": "chess-data-analysis/1.0 raipex.dev@gmail.com"}

# Verificar perfil do jogador
url_perfil = f"https://api.chess.com/pub/player/{USERNAME}"
perfil = requests.get(url_perfil, headers=headers).json()

print(f"Jogador:  {perfil['username']}")
print(f"Rating:   {perfil.get('fide', 'não informado')}")
print(f"Membro desde: {datetime.fromtimestamp(perfil['joined']).strftime('%d/%m/%Y')}")
```

```python
# Ver todos os meses disponíveis
url_arquivos = f"https://api.chess.com/pub/player/{USERNAME}/games/archives"
arquivos = requests.get(url_arquivos, headers=headers).json()

meses = arquivos["archives"]
print(f"Total de meses com partidas: {len(meses)}")
print("Últimos 5 meses:")
for m in meses[-5:]:
    print(f"  {m}")
```

```python
# Coletar partidas de um mês específico — começa pelo mais recente
url_mes = meses[-1]
response = requests.get(url_mes, headers=headers)
partidas = response.json()["games"]

print(f"Partidas no mês: {len(partidas)}")
print("\nPrimeira partida:")
print(json.dumps(partidas[0], indent=2))
```
